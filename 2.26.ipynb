```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
student_data_math = pd.read_csv("chap3/student-mat.csv", sep=";")
print(student_data_math.head())
```

```python
# 練習問題7-1
print(student_data_math["reason"].value_counts())
labels = ["course", "home", "reputation", "other"]
sizes = student_data_math["reason"].value_counts()

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
```

```python
# 別
print(student_data_math.groupby("reason").size())
student_data_math.groupby("reason").size().plot(kind="pie", autopct="%1.1f%%", startangle=90)
```

```python
# 練習問題7-2
data = student_data_math.groupby("higher")["G3"].mean()
print(data)
x = [1, 2]
y = data
plt.bar(x, y, align="center")
plt.xticks(x, ["no", "yes"])
plt.xlabel("higher")
plt.ylabel("G3")
```

```python
# 別
print(student_data_math.groupby("higher")["G3"].mean())
student_data_math.groupby("higher")["G3"].mean().plot(kind="bar")
plt.xlabel("higher")
plt.ylabel("G3")
```

```python
# 練習問題7-3
data2 = student_data_math.groupby("traveltime")["G3"].mean()
print(data2)
x = np.arange(4)
y = data2
plt.barh(x, y, align="center")
plt.yticks(x, ["1", "2", "3", "4"])
plt.ylabel("traveltime")
plt.xlabel("G3")
```

```python
# 別
print(student_data_math.groupby("traveltime")["G3"].mean())
student_data_math.groupby("traveltime")["G3"].mean().plot(kind="barh")
plt.xlabel("G3")
```
