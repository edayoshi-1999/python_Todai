```python
# 1章総合問題
# 1-1
n_list = range(2, 10 + 1)
for i in range(2, int(10 ** 0.5) + 1):
    n_list = [x for x in n_list if(x == i or x % i != 0)]

for j in n_list:
    print(j)
```

```python
# 1-2
def calc_prime_num(N):
    n_list = range(2, N + 1)
    for i in range(2, int(N ** 0.5) + 1):
        n_list = [x for x in n_list if(x == i or x % i != 0)]

    for j in n_list:
        print(j)

calc_prime_num(10)
```

```python
import scipy.linalg as linalg
from scipy.optimize import minimize_scalar
```

```python
matrix1 = np.array([[3, 2], [0, 2]])
print(matrix1)
print(linalg.det(matrix1))
matrix2 = np.array([[2, 0, 1], [-1, 1, 0], [-2, 3, 4]])
print(matrix2)
print(linalg.det(matrix2))
```

```python
matrix3 = np.array([
    [2, 1, 1],
    [1, 1, 1],
    [-2, 0, 1]
])
print(matrix3)
print(linalg.inv(matrix3))
print(matrix3.dot(linalg.inv(matrix3)))
```

```python
matrix4 = np.array([
    [2, 1, 1],
    [1, 2, 1],
    [1, 1, 2]
])
eig_value, eig_vector = linalg.eig(matrix4)
print(eig_value)
print(eig_vector)
```

```python
from scipy.optimize import newton
def my_function(x):
    return (x**2 + 2*x + 1)
print(newton(my_function, 0))

print(minimize_scalar(my_function, method="Brent"))
```
