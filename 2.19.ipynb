```python
import numpy as np
import scipy as sp
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 練習問題4-7
for df, c in zip([5, 25, 50], "bgr"):
    x = np.random.chisquare(df, 1000)
    plt.hist(x, 20, color=c)
    plt.grid(True)
```

```python
# 練習問題4-8
x = np.random.standard_t(100, 1000)
plt.hist(x)
plt.grid(True)
```

```python
# 練習問題4-9
for df, c in zip([(10, 30), (20, 25)], "bg"):
        x = np.random.f(df[0], df[1], 1000)
        plt.hist(x, 100, color=c)
        plt.grid(True)
```

```python
student_data_math = pd.read_csv(r"chap3\student-mat.csv", sep=";")
student_data_por = pd.read_csv(r"chap3\student-por.csv", sep=";")
student_data_merge = pd.merge(
                    student_data_math,
                    student_data_por,
                    on=["school", "sex", "age", "address", "famsize", "Pstatus", "Medu", "Fedu", "Mjob", "Fjob", "reason",
                       "nursery", "internet"],
                    suffixes=("_math", "_por"))
print("数学G1", student_data_merge.G1_math.mean())
print("ポルトガル語G1", student_data_merge.G1_por.mean())
```

```python
t, p = sp.stats.ttest_rel(student_data_merge.G1_math, student_data_merge.G1_por)
print("P値", p)
```

```python
# 練習問題4-13
t, p = sp.stats.ttest_rel(student_data_merge.G2_math, student_data_merge.G2_por)
print("G2のP値", p)
t, p = sp.stats.ttest_rel(student_data_merge.G3_math, student_data_merge.G3_por)
print("G3のP値", p)
```

```python
# 4章総合問題
t, p = sp.stats.ttest_rel(student_data_merge.absences_math, student_data_merge.absences_por)
print("absencesのP値", p)
t, p = sp.stats.ttest_rel(student_data_merge.studytime_math, student_data_merge.studytime_por)
print("studytimeのP値", p)
```
